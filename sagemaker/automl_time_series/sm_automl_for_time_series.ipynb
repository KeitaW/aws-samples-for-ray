```python
import sagemaker
from sagemaker.pytorch import PyTorch
import boto3
```

```python
s3_client = boto3.client("s3")
sess = sagemaker.session.Session()
role = sagemaker.get_execution_role()
bucket = sess.default_bucket()
key_prefix = "vjain-ray-data"
```

```python
# !wget --no-check-certificate --no-proxy 'http://anonymous@m5-benchmarks.s3.amazonaws.com/data/train/target.parquet'
```

```python
input_data = sess.upload_data("target.parquet", bucket, key_prefix=f"{key_prefix}/input")
```

```python
subnets=None
security_group_ids=None
```

```python
# Pytorch Image is used to enable distributed GPU training
estimator_gpu = PyTorch(
    source_dir="src-3.0.0-dev",
    entry_point="train_automl_for_time_series.py",
    subnets=subnets,
    security_group_ids=security_group_ids,
    role=role,
    instance_count=1, 
    instance_type="ml.g4dn.xlarge",
    framework_version="1.13",
    py_version="py39",
    keep_alive_period_in_seconds=1800
)
```

```python
estimator_gpu.fit({"train": input_data})
```
